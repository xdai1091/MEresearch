### A new way to make time-series query

```python
# Import some necessary packages
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

import scipy.stats as stats
import seaborn as sns
sns.set_style("white")

import time
import timeit

import copy
```

### Generate Time-series
Each time-series contain 50 points

```python
sample_size = 1000000
per_seg = 50
seg_num = sample_size/per_seg
bin_arr = np.arange(-10,20,1)
```

```python
def generate_time_series(num,per_number,db = False):
    """
    This function will help us generate a time series, as we just need to 
    input how many of them there are
    Input: 
        num, the total number of points to generate
        per_number: the number of points per time-series
    """
    # Generate data 
    
    # setting hyperprameters 
    h_alpha=4
    h_beta=1

    sigma_square=stats.invgamma.rvs(size=1,a=h_alpha,scale=h_beta)[0]
    print "sigma_square is:", sigma_square

    # Generate observations
    sample_size=num
    y=np.array([0.0 for i in range(sample_size)])
    noise=np.array(stats.norm.rvs(size=sample_size,loc=0,scale=np.sqrt(sigma_square)))

    for i in range(1,sample_size):
        alpha= np.random.uniform(0.1,1,size = 1)
        y[i]=alpha*y[(i-1)]+noise[i]
    
    if(db == False):
        return y
    
    # Separate points to get time-series 
    time_series = {}
    
    for i in range(num/per_number):
        time_series[i] = y[i*per_number:per_number + i*per_number]
    
    
    
    return y ,time_series
```

```python
#generate 1 millions points and each time-series have 50 points
all_data, time_series_db =  generate_time_series(1000000,50, True)
```

```python
# plot the first time_series
plt.plot(time_series_db[0])
plt.title("The first time-series in our db")
```

### Single Value Indexing Postings

```python
def sig_pos(all_data, bin_arr = np.arange(-10,20,1)):
    # Discretize the time-series 
    dis_y = np.digitize(all_data,bin_arr)
    # loop through each time series point 
    # postings have [the segment #]: [which time series contain such segment + the place where this segment is]
    # i: the ith time-series
    # j: the jth segment, 1 - 10
    postings = {}
    for i in range(seg_num):
        for j in range(per_seg):
            if(dis_y[j + per_seg*i] in postings):
                postings[dis_y[j+per_seg*i]].append((i,j))
            else:
                postings[dis_y[j+per_seg*i]] = [(i,j)]
    return postings
```

### Pairs of Values Indexing Postings

```python
def dob_pos(all_data,  bin_arr = np.arange(-10,20,1)):
    # Discretize the time-series 
    dis_y = np.digitize(all_data,bin_arr)
    doub_pos = {}
    for i in range(seg_num):
        for j in range(per_seg - 1):
            if((dis_y[j + per_seg*i],dis_y[j + 1 + per_seg*i]) in doub_pos):
                doub_pos[((dis_y[j + per_seg*i],dis_y[j + 1 + per_seg*i]))].append((i,j))
            else:
                doub_pos[((dis_y[j + per_seg*i],dis_y[j + 1 + per_seg*i]))] = [(i,j)]
    return doub_pos
```

```python
#Initialize two postings
single_pos = sig_pos(all_data)
doub_pos = dob_pos(all_data)
```

### Find Values by Traditional Way

```python
# use the interted indexing to find the similar shape
# Use Euclidine to find the smallest distance among those time-series which have the highest frequencies
def findByIdxL2(search,postings,time_sereis_db,length_series = per_seg):
    result = []
    miss = 0
    for j in xrange(length_series):
        if(miss > (int(length_series/2))):
            print "No Similar Result Found"
            return 0
        elif (search[j] in postings):
            for item in postings[search[j]]:
                if(item[1] == j):
                    result.append(item)
        else:
            miss += 1
    print np.bincount([item[0] for item in result])
    freq_list = np.bincount([item[0] for item in result])
    #find the maximum values' indexes 
    max_num = max(freq_list)
    max_index = [i for i, j in enumerate(freq_list) if j == max_num]
    dist = [np.sum(np.square(np.subtract(search,time_sereis_db[j]))) for j in max_index]
    return max_index[np.argmin(dist)]   
```

### Find Values by Pair of Values

```python
def findByPair(query,postings,time_series_db,per_seg):
    result = []
    miss = 0
    for j in xrange(per_seg-1):
        if(miss > (int(per_seg/4))):
            print "No Similar Result Found"
            return 0
        elif((query[j],query[j+1]) in postings):
            for item in postings[(query[j],query[j+1])]:
                if(item[1] == j or item[1] == (j-1) or item[1] == (j+1)):
                    result.append(item)
        else:
            miss += 1
    print np.bincount([item[0] for item in result])
    freq_list = np.bincount([item[0] for item in result])
    #find the maximum values' indexes 
    max_num = max(freq_list)
    max_index = [i for i, j in enumerate(freq_list) if j == max_num]
    dist = [np.sum(np.square(np.subtract(query,time_series_db[j]))) for j in max_index]
    return max_index[np.argmin(dist)]   
```

```python
np.random.seed(10)
#generate a new time series
query = generate_time_series(50,50,False)
# query = all_data[50:100]
# discretize this query
dis_q = np.digitize(query, bin_arr)
matched_time = findByPair(dis_q,doub_pos,time_series_db,per_seg)
plt.plot(query)
plt.plot(time_series_db[matched_time],c='y')
plt.plot(time_series_db[findByIdxL2(dis_q,single_pos,time_series_db,per_seg)], c = 'r')
```

### Performance Comparison
The two ways I discussed above shall have the same accuracy yet different performances. The following section tries to compare these two different approaches' performances

```python
#starting with data points 1 millions, 1.5 millision to 10 million to 15 millions
data_size = 1000000
query_dict = {}
pair_v_dict = {}
single_v_dict = {}
time_spent_pair = []
time_spent_single = []
single_v_acc = []
pair_v_acc = []
for i in xrange(20):
    print "Starting:",i
    np.random.seed(10)
    query = generate_time_series(50,50,False)
    query_dict[i] = query
    dis_q = np.digitize(query, bin_arr)
    test_data, test_series_db =  generate_time_series(data_size + i * 500000,50,True)
    single_pos = sig_pos(test_data)
    doub_pos = dob_pos(test_data)
    t0 = time.clock()
    pair_v = findByPair(dis_q,doub_pos,time_series_db,per_seg)
    time_spent_pair.append(time.clock() - t0)
    # calculate accuracy and add to the list
    pair_v_acc.append(np.sqrt(np.sum(np.square(np.subtract(time_series_db[pair_v],query)))))
    pair_v_dict[i] = time_series_db[pair_v]
    t0 = time.clock()
    single_v = findByIdxL2(dis_q,single_pos,time_series_db,per_seg)
    time_spent_single.append(time.clock()-t0)
    # calculate accuracy and add to the list
    single_v_acc.append(np.sqrt(np.sum(np.square(np.subtract(time_series_db[single_v],query)))))
    single_v_dict[i] = time_series_db[single_v]
```

```python
plt.plot(time_spent_single,c='y',alpha = 0.5)
plt.plot(time_spent_pair,c='b',alpha = 0.5)
plt.title("Time spent on both ways")
plt.xlabel("Iterations")
plt.ylabel("Time spent(microsecond)")
```

```python
plt.plot(single_v_acc,c='y',alpha = 0.5)
plt.plot(pair_v_acc, c='b',alpha = 0.5)
```

### Accuracy Measurement (Single/Pair)
In order to calculate the accuracy. I decided to use the following formulas.
$$accuracy = -\frac{1}{n}\sqrt{||\bar{x_i} - x ||_2^2}$$
where $\bar{x}$ is the neighborhood we found, and $x$ is the original time-series

```python
print "The accuracy for single:", -np.mean(single_v_acc)
print "The accuracy for pair",-np.mean(pair_v_acc)
```

## Adaptive Indexing (Tree-based)
The idea of this part is to construct a tree, which can promote the cardinalities during the time when we add more time-series added(which results in more accuracy result). Each node shall only contain certain number of similar time-series(threshold). When it reaches this limit, the cardinalities shall be promoted (the node should be split). I will use the hash table to store the time-series information. For example, a time-series having {6_8, 6_8, 3_8, 0_8} should be represented as the key "6.8_6.8_3.8_0.8". and the value is the list of similar time-series. The node will be split when there are more time series than the threshold. 


```python
def mapBin(card,min_val,max_val):
    assert card > 1, "Cardinality should be at least 1"
    res = []
    diff = max_val - min_val
    gap = diff / card
    for x in xrange(card-1):
        res.append(min_val+gap * (x+1))
    assert len(res) == (card-1), "The length of result is wrong!"
    return res
```
